```python
train_dir = '../input/dog-breed-identification/train'
test_dir ='../input/dog-breed-identification/test'
```

```python
def append_ext(fn):
    return fn+".jpg"
traindf = pd.read_csv('../input/dog-breed-identification/labels.csv',dtype=str)
testdf = pd.read_csv('../input/dog-breed-identification/sample_submission.csv',dtype=str)
traindf["id"] = traindf["id"].apply(append_ext)
testdf["id"] = testdf["id"].apply(append_ext)
```

```python
train_datagen=ImageDataGenerator(rescale=1./255.,
                                 rotation_range = 20,
                                 brightness_range=[0.2,1.0],
                                 width_shift_range = 0.2,
                                 height_shift_range = 0.2,
                                 shear_range = 0.2,
                                 zoom_range = [0.7,1],
                                 horizontal_flip = True,
                                 validation_split=0.02 
                                 )
```

```python
BATCH_SIZE = 32
```

```python
image_size=(299,299)
```

```python
train_generator=train_datagen.flow_from_dataframe(dataframe=traindf,
                                                  directory=train_dir,
                                                  x_col="id",
                                                  y_col="breed",
                                                  subset="training",
                                                  batch_size=BATCH_SIZE,
                                                  seed=42,
                                                  shuffle=True,
                                                  class_mode="categorical",
                                                  target_size=image_size,
                                                  color_mode="rgb" 
                                                 )
```

```python
def show_grid(image_list,nrows,ncols,figsize=(10,10),showaxis='off'):
    
    image_list = [image_list[i,:,:,:] for i in range(image_list.shape[0])]
    fig = plt.figure(None, figsize,frameon=False)
    grid = ImageGrid(fig, 111,  # similar to subplot(111)
                     nrows_ncols=(nrows, ncols),  # creates 2x2 grid of axes
                     axes_pad=0.3,  # pad between axes in inch.
                     share_all=True,
                     )
    for i in range(nrows*ncols):
        ax = grid[i]
        ax.imshow(image_list[i],cmap='Greys_r')  # The AxesGrid object work as a list of axes.
        ax.axis('off')
        
```

```python
show_grid(x,3,8,figsize=(25,25))
```

```python
valid_generator=train_datagen.flow_from_dataframe(dataframe=traindf,
                                                  directory=train_dir,
                                                  x_col="id",
                                                  y_col="breed",
                                                  subset="validation",
                                                  batch_size=BATCH_SIZE,
                                                  seed=42,
                                                  shuffle=True,
                                                  class_mode="categorical",
                                                  target_size=image_size,
                                                  color_mode="rgb")
```

```python
shape=(299,299,3)
```

```python
pretrained_model = tf.keras.applications.InceptionResNetV2(weights='imagenet',include_top=False ,input_shape=shape)
pretrained_model.trainable = False   
```

```python
model = tf.keras.Sequential([ 
        pretrained_model,   
        tf.keras.layers.GlobalAveragePooling2D(),
        #tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(120, activation='softmax')
    ])
```

```python
opt=tf.keras.optimizers.SGD(lr=1e-3, momentum=0.9)
model.compile(optimizer = opt ,
              loss="categorical_crossentropy",
              metrics=["accuracy"])
model.summary()
```

```python
early = tf.keras.callbacks.EarlyStopping(patience=2,min_delta=0.001,restore_best_weights=True)
```

```python
STEP_SIZE_TRAIN = train_generator.n//train_generator.batch_size
STEP_SIZE_VALID = valid_generator.n//valid_generator.batch_size

history = model.fit(train_generator,steps_per_epoch=STEP_SIZE_TRAIN,validation_data=valid_generator,validation_steps=STEP_SIZE_VALID,epochs=50,
                    callbacks=[early],)
```

```python
def display_training_curves(training, validation, title, subplot):
    if subplot%10==1: # set up the subplots on the first call
        plt.subplots(figsize=(10,10), facecolor='#F0F0F0')
        plt.tight_layout()
    ax = plt.subplot(subplot)
    ax.set_facecolor('#F8F8F8')
    ax.plot(training)
    ax.plot(validation)
    ax.set_title('MODEL '+ title)
    ax.set_ylabel(title)
    ax.set_xlabel('epoch')
    ax.legend(['train', 'valid.'])
```

```python
display_training_curves(history.history['loss'],history.history['val_loss'],'LOSS',211,)
display_training_curves(history.history['accuracy'],history.history['val_accuracy'],'ACCURACY',212,)
```

```python
score = model.evaluate(valid_generator,batch_size=32)
print("Accuracy: {:.2f}%".format(score[1] * 100)) 
print("Loss: ",score[0])
```

```python
test_datagen=ImageDataGenerator(rescale=1./255.)

test_generator=test_datagen.flow_from_dataframe(
dataframe=testdf,
directory=test_dir,
x_col="id",
y_col=None,
batch_size=BATCH_SIZE,
seed=42,
shuffle=False,
class_mode=None,
target_size=image_size,
color_mode="rgb")
```

```python
pred=model.predict(test_generator)
```
